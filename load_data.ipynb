# Define imports and helper functions

```python
import os
import shutil
import numpy as np
import cv2
```

```python
#helper function
def load_images_from_folder(folder_path, target_size=(80, 80)):
    if 'images' in locals() and 'image_files' in locals():
      print("Images already loaded.")
      return images, image_files
    images = []
    image_files = []
    for filename in os.listdir(folder_path):
        img_path = os.path.join(folder_path, filename)
        if os.path.isfile(img_path):
            #load the image, resize it, and convert to array
            img = cv2.imread(img_path)
            img = cv2.resize(img, target_size)
            images.append(img)
            image_files.append(filename)
    return np.array(images), np.array(image_files)
```

```python
def create_dataset(image_folder_path):
    images, image_files = load_images_from_folder(image_folder_path)
    np.savez(f'{path}/dataset.npz', images=images, image_files=np.array(image_files))
```

# Create the unlabeled dataset

```python
#define the paths for the ship and no-ship folders
ship_folder = os.path.join(path, 'ship')
no_ship_folder = os.path.join(path, 'no-ship')
unlabeled_dataset_folder = os.path.join(path, 'unlabeled_dataset')


#create a new folder for the unlabeled dataset if it doesn't exist yet
if not os.path.exists(unlabeled_dataset_folder):
    os.makedirs(unlabeled_dataset_folder)

#copy ship images to the unlabeled dataset folder
for file_name in os.listdir(ship_folder):
    file_path = os.path.join(ship_folder, file_name)
    if os.path.isfile(file_path):
        shutil.copy(file_path, unlabeled_dataset_folder)

#copy no-ship images to the unlabeled dataset folder
for file_name in os.listdir(no_ship_folder):
    file_path = os.path.join(no_ship_folder, file_name)
    if os.path.isfile(file_path):
        shutil.copy(file_path, unlabeled_dataset_folder)
```

```python
#count the images in the ship and no-ship folders
ship_count = len([name for name in os.listdir(ship_folder) if os.path.isfile(os.path.join(ship_folder, name))])
no_ship_count = len([name for name in os.listdir(no_ship_folder) if os.path.isfile(os.path.join(no_ship_folder, name))])

#count the total number of images in the unlabeled dataset folder
unlabeled_dataset_count = len([name for name in os.listdir(unlabeled_dataset_folder) if os.path.isfile(os.path.join(unlabeled_dataset_folder, name))])

#output the counts
print(f'Total number of ship images: {ship_count}') # should be 1000
print(f'Total number of no-ship images: {no_ship_count}') # should be 3000
print(f'Total number of images in the new unlabeled dataset: {unlabeled_dataset_count}') # should be 4000
```
